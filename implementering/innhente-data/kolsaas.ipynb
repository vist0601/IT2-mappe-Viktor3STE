```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[1][0] #gir meg den f√∏rste tiden (0)
```

```python
x = []
y = []


for linje in tur[1:]:
    tid = float(linje[0])
    hoyde = float(linje[3])
    x.append(tid)
    y.append(hoyde)

plt.plot(x, y)
plt.show()




```
